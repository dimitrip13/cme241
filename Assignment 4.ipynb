```python
from dataclasses import dataclass
from typing import Tuple, Dict, List, Optional, Iterable, Mapping
from rl.markov_decision_process import FiniteMarkovDecisionProcess
from rl.markov_decision_process import FinitePolicy, StateActionMapping
from rl.markov_process import FiniteMarkovProcess, FiniteMarkovRewardProcess, Transition
from rl.distribution import Categorical, Constant, Choose, FiniteDistribution
from rl.dynamic_programming import *
import pprint as pp
import numpy as np
import itertools
import time
```

```python
@dataclass(frozen=True)
class Lilypad:
    state: int
    
    # Was having weird equality problems when indexing into a dict with the state as a key 
    # so I added this in.  It overloads the default == behavior
    def __eq__(self, other):
        if self.state == other.state:
            return True
        else:
            return False

# Maps {States -> Actions}
TransitionMapping = StateActionMapping[Lilypad, int]


# Models the Lilypads game as an MDP
@dataclass
class LilypadMDP(FiniteMarkovDecisionProcess[Lilypad, int]):    
    win: int = 10
    lose: int = 0
    actionss: tuple = (0, 1) # Can't make this a list
    #ingredients: List = field(default_factory=lambda: ['dow', 'tomatoes'])
    
    def __post_init__(self):
#         self.loopactions = list(self.actions)
        self.statespace = [state for state in range(self.lose, self.win + 1)]
        self.actionss = list(self.actionss)
        self.state_actions = self.make_state_action_space()
        self.n = self.win
        super().__init__(self.get_action_transition_reward_map())
        
    def make_state_action_space(self):
        arr = []
        for state in self.statespace:
            if state == self.lose:
                arr.append((Lilypad(state), -1))
            elif state == self.win:
                arr.append((Lilypad(state), 1))       
            else:
                arr.append((Lilypad(state), 0))
        return arr
    
    def get_action_transition_reward_map(self) -> TransitionMapping:
        # {State -> {Action -> Distribution(Next State, Reward)}}
        d: Dict[Lilypad, Dict[int, Optional[Tuple[Lilypad, float]]]] = {}
        
        # Loops over each state
        for state in self.statespace:
            # {Action -> Distribution(Next State, Reward)}
            d1: Dict[int, Optional[Tuple[Lilypad, float]]] = {}
            
            # Terminal state
            if state == self.lose:
                d1 = None
                
            # Terminal state
            elif state == self.win:
                d1 = None
    
            # Non terminal states
            else:
                # Loops over actions and possbile next states
                for action in self.actionss:
        
                    # Action A
                    if action == 0:
                        d1[action] = Categorical({self.state_actions[state - 1] : state/self.n, 
                                            self.state_actions[state + 1] : (self.n - state)/self.n})
                    # Action B
                    elif action == 1:
                        d1[action] = Choose(self.state_actions)
                    
                    else: 
                        raise "Bruh what is going on here"
        
            d[Lilypad(state)] = d1

        return d
    
    
# A basic policy where all actions occur with equal probability
# The same actions can be taken everywhere except terminal states
# Policy Maps   Dict[State -> Distribution(Actions)]
def make_basic_policy(statespace, actionslist) -> Mapping[Lilypad, Optional[FiniteDistribution[int]]]:
    d = {}  # Set empty dict
    equal_prob = 1/len(actionslist)   # Normalize Probability
    
    # Loops over each state
    for state in statespace:
        inner_dict = {}           # Creates empty inner dict
        
        # Loops over each action
        for action in actionslist:
            inner_dict[action] = equal_prob
        
        # Sets terminal states to None
        if (state == lose) or (state == win):
            d[Lilypad(state)] = None
        # Sets state to map to created distribution of actions
        else: 
            d[Lilypad(state)] = Categorical(inner_dict)
            
    return d


# Checks each combination of states and actions and returns the optimal policy with the 
# largest value function as well as the value functions
def get_brute_force_optimal_policy(lily: LilypadMDP, gamma: int):
    # Gets common variables from MDP
    lose = lily.lose
    win = lily.win
    actionlist = lily.actionss
    
    maxvec = np.zeros((1,win-1))        # Sets value function vector to zeros
    best_policy = None                  # Inits the best policy as None
    
    # Gets a list of non-terminal states and their number
    non_terminal_states_list = [state for state in range(lose+1, win)]
    num_nts = len(non_terminal_states_list) # Number of non-terminal states
    
    # Gets an interator for each combination of actions for each non-terminal state
    # Ex 2 non-terminal states, 3 actions -> (0,0),(0,1),(0,2),(1,0),(1,1),(1,2)
    combos = itertools.product(actionlist, repeat=num_nts)
    #  print(list(iter(combos)))

    # Loops over each combination of actions
    for c in combos:
        # Initializes an empty dict of the policy map
        policy_map = {}
        
        # Loops over the number of non-terminal states
        for i in range(num_nts):
            state = non_terminal_states_list[i]  # Get the state value
            
            # Adds each state and action to the policy
            policy_map[Lilypad(state)] = Constant(c[i])
        
        # Add the terminal states
        policy_map[Lilypad(win)] = None
        policy_map[Lilypad(lose)] = None
        
        # Make the finite polciy from the policy map
        policy: FinitePolicy[Lilypad, Optional[FiniteDistribution[int]]] = FinitePolicy(policy_map)

        # Make the Markov Reward Process
        mrp: FiniteMarkovRewardProcess[Lilypad] = lily.apply_finite_policy(policy)
        
        # Get the value function vector of the policy
        v_func_vec = mrp.get_value_function_vec(gamma)
         
        # If the new vector has a better value function sum, then update the max and the policy
        if np.sum(v_func_vec) > np.sum(maxvec):
            maxvec = v_func_vec
            best_policy = policy
                  
    return maxvec, best_policy
  
    
# Generates the best policy with policy iteration and returns the policy when it converges to 
# the given tolerance metric
def policy_iteration_get_best_policy(policy_iter, tolerance=10e-15):
    first_time = True
    while True:
        # Gets the next iteration of the value function and policy
        value_function, policy = next(policy_iter)
        
        value_function_vec = []   # Inits empty list
        
        # Loops over each value to get the value function into a list
        # Step is iteration count, and stateval is a tuple of (state, utility (V(state)))
        for step, (state, val) in enumerate(value_function.items()):  
            value_function_vec.append(val)
        value_function_vec = np.array(value_function_vec) # Turns list into np array
        
        # Creates an array of infinity to compare to for the first time
        # This is placed inside the loop to make it the correct size
        if first_time:
            prev = np.full((1, len(value_function_vec)), fill_value=np.Inf)
            # Toggles the flag false
            first_time = False
        
        # If norm between arrays < the difference tolerance, 
        # quits out returning the optimal policy and value function vector
        if np.linalg.norm(value_function_vec - prev) < tolerance:
            return policy, value_function_vec
        # Else, sets the current value function to be the previous for the next iteration
        else:
            prev = value_function_vec
            
            
# Generates the optimal value funtion vector with value iterations
# Given an value function iterate, returns the optimal value funtion vector when it 
# converges to the given tolerance metric
def value_iteration_get_value_functions(value_iter, tolerance=10e-15):
    first_time = True
    while True:
        # Gets the next iteration of the value function and policy
        value_function= next(value_iter)

        value_function_vec = []   # Inits empty list
        
        # Loops over each value to get the value function into a list
        # Step is iteration count, and stateval is a tuple of (state, utility (V(state)))
        for step, (state, val) in enumerate(value_function.items()):  
            value_function_vec.append(val)
        value_function_vec = np.array(value_function_vec) # Turns list into np array
        
        # Creates an array of infinity to compare to for the first time
        # This is placed inside the loop to make it the correct size
        if first_time:
            prev = np.full((1, len(value_function_vec)), fill_value=np.Inf)
            # Toggles the flag false
            first_time = False
        
        # If norm between arrays < the difference tolerance, 
        # quits out returning the value function vector
        if np.linalg.norm(value_function_vec - prev) < tolerance:
            return value_function_vec 
        # Else, sets the current value function to be the previous for the next iteration
        else:
            prev = value_function_vec

            
            
# Takes a value function with type V[S] and converts it into a numpy array of size (n,)
def value_function_to_vector(vf):
    value_function_vec = []   # Inits empty list
        
    # Loops over each value to get the value function into a list
    # Step is iteration count, and stateval is a tuple of (state, utility (V(state)))
    for step, (state, val) in enumerate(vf.items()):  
        value_function_vec.append(val)
    value_function_vec = np.array(value_function_vec) # Turns list into np array
    return value_function_vec            
```

```python
from dataclasses import dataclass
from typing import Tuple, Dict
from rl.markov_decision_process import FiniteMarkovDecisionProcess
from rl.markov_decision_process import FinitePolicy, StateActionMapping
from rl.markov_process import FiniteMarkovProcess, FiniteMarkovRewardProcess
from rl.distribution import Categorical, Constant
from scipy.stats import poisson


@dataclass(frozen=True)
class InventoryState:
    inventory1: int
    inventory2: int
    on_order1: int
    on_order2: int
        
InvOrderMapping = StateActionMapping[InventoryState, int]        
        
    
class FunkyInventoryMDPCap(FiniteMarkovDecisionProcess[InventoryState, int]):

    def __init__(
        self,
        capacity1: int,
        poisson_lambda1: float,
        holding_cost1: float,
        stockout_cost1: float,
        
        capacity2: int,
        poisson_lambda2: float,
        holding_cost2: float,
        stockout_cost2: float,
        
        supply_cost: float,
        transfer_cost: float
        
    ):
        self.capacity1: int = capacity1
        self.poisson_lambda1: float = poisson_lambda1
        self.holding_cost1: float = holding_cost1
        self.stockout_cost1: float = stockout_cost1
            
        self.capacity2: int = capacity2
        self.poisson_lambda2: float = poisson_lambda2
        self.holding_cost2: float = holding_cost2
        self.stockout_cost2: float = stockout_cost2
            
        self.supply_cost: float = supply_cost
        self.transfer_cost: float = transfer_cost

        self.poisson_distr1 = poisson(poisson_lambda1)
        self.poisson_distr2 = poisson(poisson_lambda2)
        
        super().__init__(self.get_action_transition_reward_map())

    def get_action_transition_reward_map(self) -> InvOrderMapping:
        # {State -> {Action -> Distribution(Next State, Reward)}}
        d: Dict[InventoryState, Dict[str, Categorical[Tuple[InventoryState, float]]]] = {}
        # Loop over all possible store 1 inventory
        for alpha_1 in range(self.capacity1 + 1):
            # Loop over all possible store 1 orders
            for beta_1 in range(self.capacity1 + 1 - alpha_1):
                # Loop over all possible store 2 inventory
                for alpha_2 in range(self.capacity2 + 1):
                    # Loop over all possible store 2 orders
                    for beta_2 in range(self.capacity2 + 1 - alpha_2):
                        state = InventoryState(alpha_1, alpha_2, beta_1, beta_2)
                        
                        holding_cost_reward_1 = self.holding_cost1 * alpha_1
                        holding_cost_reward_2 = self.holding_cost2 * alpha_2
                        base_reward = holding_cost_reward_1 + holding_cost_reward_2
                        
                        # {Action -> Distribution(Next State, Reward)}
                        d1: Dict[str, Categorical[Tuple[InventoryState, float]]] = {}
                            
                        # Tuple[NextState, Reward] -> Transition Probability
                        sr_probs_dict: Dict[Tuple[InventoryState, float], float] = {}
                            
                        # Loop over all possible orders from store 1    
                        for order_1 in range(self.capacity1 - state.inventory1 + 1):
                            # Loop over all possible orders from store 2    
                            for order_2 in range(self.capacity2 - state.inventory2 + 1):
                                # Loop over the possible store transfer amount (positive or negative)
                                for transfer_amount in range(-(self.capacity1 - state.inventory1 - order_1), self.capacity2 - state.inventory2 - order_2):  
                                    
                                    # Check to see if we need to apply a transfer cost
                                    if transfer_amount == 0:
                                        transfer_cost = 0
                                    else:
                                        transfer_cost = self.transfer_cost

                                    # Check to see if we need to apply a supply cost
                                    if order_1 > 0 or order_2 > 0:
                                        supply_cost = 2*self.supply_cost
                                    elif order_1 > 0:
                                        supply_cost = self.supply_cost
                                    elif order_2 > 0:
                                        supply_cost = self.supply_cost
                                    else:
                                        supply_cost = 0
                                                         
                                    base_reward_with_actions = base_reward + supply_cost + transfer_cost
                                    
                                    # Create action of a sting (so it's readable when printed)
                                    action_str = f"Store 1 buy: {order_1} \t Store 2 buy: {order_2} \t Transfer {transfer_amount} from A to B"

                                    # Loop over all possible demand in store 1
                                    for purchase_1 in range(state.inventory1 + 1):
                                        # Loop over all possible demand in store 2
                                        for purchase_2 in range(state.inventory2 + 1):
                                            # Get probability of demand for both stores
                                            prob = self.poisson_distr1.pmf(purchase_1) * self.poisson_distr2.pmf(purchase_2)
                            
                                            # Create next state
                                            next_state = InventoryState(max(alpha_1-transfer_amount-purchase_1,0), max(alpha_2+transfer_amount-purchase_2,0), beta_1+order_1, beta_2+order_2)
                                            
                                            # Add to probability dict # Tuple[NextState, Reward] -> Transition Probability
                                            sr_probs_dict[(next_state, base_reward_with_actions)] = prob
                                    
                                    # Compute stockout cost for store 2
                                    for purchase_1 in range(state.inventory1 + 1):
                                        # Probability of selling below their demand (to calculate stockout cost ie demand when we don't have supply)
                                        prob = self.poisson_distr1.pmf(purchase_1) * (1 - self.poisson_distr2.cdf(state.inventory2))
                                        next_state = InventoryState(alpha_1 - purchase_1, 0, order_1, order_2)
                                        store_2_stockout_reward = base_reward_with_actions - self.stockout_cost2 * \
                                            (self.poisson_lambda2 * (1 - self.poisson_distr2.cdf(state.inventory2 - 1)) - (state.inventory2 * (1 - self.poisson_distr2.cdf(state.inventory2))))             
                                        sr_probs_dict[(next_state, store_2_stockout_reward)] = prob
                                        
                                    # Compute stockout cost for store 1
                                    for purchase_2 in range(state.inventory2 + 1):
                                        # Probability of selling below their demand (to calculate stockout cost ie demand when we don't have supply)
                                        prob = self.poisson_distr2.pmf(purchase_2) * (1 - self.poisson_distr1.cdf(state.inventory1))
                                        next_state = InventoryState(0, alpha_2 - purchase_2, order_1, order_2)
                                        store_1_stockout_reward = base_reward_with_actions - self.stockout_cost1 * \
                                            (self.poisson_lambda1 * (1 - self.poisson_distr1.cdf(state.inventory1 - 1)) - (state.inventory1 * (1 - self.poisson_distr1.cdf(state.inventory1))))             
                                        sr_probs_dict[(next_state, store_1_stockout_reward)] = prob
                                    
                                    # This makes the data read rate to crash ... :(
                                    # Stockout cost for both store 1 and store 2
#                                     prob1 = self.poisson_distr1.cdf(state.inventory1 - 1)
#                                     prob2 = self.poisson_distr2.cdf(state.inventory2 - 1)
#                                     next_state = InventoryState(0, 0, order_1, order_2)
#                                     cost1 = self.stockout_cost1 * \
#                                             (self.poisson_lambda1 * (1 - self.poisson_distr1.cdf(state.inventory1 - 1)) - (state.inventory1 * (1 - self.poisson_distr1.cdf(state.inventory1))))  
#                                     cost2 = self.stockout_cost2 * \
#                                             (self.poisson_lambda2 * (1 - self.poisson_distr2.cdf(state.inventory2 - 1)) - (state.inventory2 * (1 - self.poisson_distr2.cdf(state.inventory2))))      
#                                     double_stockout_reward = base_reward_with_actions - cost1 - cost2
#                                     sr_probs_dict[(next_state, double_stockout_reward)] = prob1 * prob2
                                      
                                        
                                        
                                    d1[action_str] = Categorical(sr_probs_dict)
                        d[state] = d1       
        return d
```
